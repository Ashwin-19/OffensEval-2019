```python
from sklearn.manifold import TSNE
import plotly.express as px
import pandas as pd
import numpy as np

import joblib
```

```python
train_data = pd.read_csv('subtask_A_preprocessed.csv',index_col='id')
```

```python
X_train = train_data.drop(columns=['subtask_a'])
y_train = train_data['subtask_a']
```

```python
tsne = TSNE(n_components=2)
```

```python
embedding_2d = tsne.fit_transform(X_train)
```

```python
fe_df = pd.DataFrame()
```

```python
fe_df['x1'] = embedding_2d[:,0]
fe_df['x2'] = embedding_2d[:,1]
fe_df['y'] = y_train.tolist()
```

```python
fig = px.scatter(fe_df, x="x1", y="x2", color="y")
fig.update_layout(title_text='T-SNE Scatter Plot of Extracted Features')

fig.show()
```

```python
X_train = joblib.load('X_train_cls')
y_train = joblib.load('y_train')
```

```python
embeddings_2d_bert = TSNE(n_components=2).fit_transform(X_train)
```

```python
bert_df = pd.DataFrame()
```

```python
fig = px.scatter(bert_df, x="x1", y="x2", color="y")
fig.update_layout(title_text='T-SNE Scatter Plot of BERT Embeddings')

fig.show()
```
